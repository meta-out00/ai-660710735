# ‡πÄ‡∏õ‡∏•‡∏µ‡πà‡∏¢‡∏ô‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏• ‡∏ä‡πà‡∏≠‡∏á ‡∏≠‡∏≤‡∏¢‡∏∏, ‡∏™‡πà‡∏ß‡∏ô‡∏™‡∏π‡∏á, ‡∏ô‡πâ‡∏≥‡∏´‡∏ô‡∏±‡∏Å, ‡∏ä‡∏±‡πà‡∏ß‡πÇ‡∏°‡∏á‡πÄ‡∏£‡∏µ‡∏¢‡∏ô_‡∏ï‡πà‡∏≠‡∏™‡∏±‡∏õ‡∏î‡∏≤‡∏´‡πå ‡∏ó‡∏µ‡πà‡πÄ‡∏õ‡πá‡∏ô NaN ‡πÄ‡∏õ‡πá‡∏ô‡∏Ñ‡πà‡∏≤‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢ & ‡∏à‡∏±‡∏î‡∏Å‡∏≤‡∏£‡∏Ñ‡πà‡∏≤‡∏ú‡∏¥‡∏î‡∏õ‡∏Å‡∏ï‡∏¥ ‡πÉ‡∏ô‡∏ä‡πà‡∏≠‡∏á ‡∏≠‡∏≤‡∏¢‡∏∏, GPA ‡πÇ‡∏î‡∏¢‡πÄ‡∏≠‡∏≤‡∏Ñ‡πà‡∏≤‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢‡∏à‡∏≤‡∏Å‡∏Å‡∏≤‡∏£‡∏à‡∏±‡∏î‡∏Å‡∏≤‡∏£‡∏Ñ‡πà‡∏≤‡∏ú‡∏¥‡∏î‡∏õ‡∏Å‡∏ï‡∏¥‡∏°‡∏≤‡πÉ‡∏™‡πà‡∏Å‡πà‡∏≠‡∏ô‡πÄ‡∏≠‡∏≤‡∏≠‡∏≠‡∏Å

```python
Q1_age = df_cleaned['‡∏≠‡∏≤‡∏¢‡∏∏'].quantile(0.25)
Q3_age = df_cleaned['‡∏≠‡∏≤‡∏¢‡∏∏'].quantile(0.75)
IQR_age = Q3_age - Q1_age

Q1_gpa = df_cleaned['GPA'].quantile(0.25)
Q3_gpa = df_cleaned['GPA'].quantile(0.75)
IQR_gpa = Q3_gpa - Q1_gpa

lower_bound_age = Q1_age - 1.5 * IQR_age
upper_bound_age = Q3_age + 1.5 * IQR_age

lower_bound_gpa = Q1_gpa - 1.5 * IQR_gpa
upper_bound_gpa = Q3_gpa + 1.5 * IQR_gpa
```

```python
df_cleaned_age_mean = df_cleaned[(df_cleaned['‡∏≠‡∏≤‡∏¢‡∏∏'] >= lower_bound_age) & (df_cleaned['‡∏≠‡∏≤‡∏¢‡∏∏'] <= upper_bound_age)]
df_cleaned_gpa_mean = df_cleaned[(df_cleaned['GPA'] >= lower_bound_gpa) & (df_cleaned['GPA'] <= upper_bound_gpa)]
```

```python
mean_age = df_cleaned_age_mean['‡∏≠‡∏≤‡∏¢‡∏∏'].mean()
df_cleaned['‡∏≠‡∏≤‡∏¢‡∏∏']=df_cleaned['‡∏≠‡∏≤‡∏¢‡∏∏'].fillna(mean_age).astype(int)

mean_gpa = df_cleaned_gpa_mean['GPA'].mean()
df_cleaned['GPA']=df_cleaned['GPA'].fillna(mean_gpa).round(2)

mean_height = df_cleaned['‡∏™‡πà‡∏ß‡∏ô‡∏™‡∏π‡∏á_cm'].mean()
df_cleaned['‡∏™‡πà‡∏ß‡∏ô‡∏™‡∏π‡∏á_cm']=df_cleaned['‡∏™‡πà‡∏ß‡∏ô‡∏™‡∏π‡∏á_cm'].fillna(mean_height).round(1)

mean_weight = df_cleaned['‡∏ô‡πâ‡∏≥‡∏´‡∏ô‡∏±‡∏Å_kg'].mean()
df_cleaned['‡∏ô‡πâ‡∏≥‡∏´‡∏ô‡∏±‡∏Å_kg']=df_cleaned['‡∏ô‡πâ‡∏≥‡∏´‡∏ô‡∏±‡∏Å_kg'].fillna(mean_weight).round(1)

mean_hour_per_week = df_cleaned['‡∏ä‡∏±‡πà‡∏ß‡πÇ‡∏°‡∏á‡πÄ‡∏£‡∏µ‡∏¢‡∏ô_‡∏ï‡πà‡∏≠‡∏™‡∏±‡∏õ‡∏î‡∏≤‡∏´‡πå'].mean()
df_cleaned['‡∏ä‡∏±‡πà‡∏ß‡πÇ‡∏°‡∏á‡πÄ‡∏£‡∏µ‡∏¢‡∏ô_‡∏ï‡πà‡∏≠‡∏™‡∏±‡∏õ‡∏î‡∏≤‡∏´‡πå']=df_cleaned['‡∏ä‡∏±‡πà‡∏ß‡πÇ‡∏°‡∏á‡πÄ‡∏£‡∏µ‡∏¢‡∏ô_‡∏ï‡πà‡∏≠‡∏™‡∏±‡∏õ‡∏î‡∏≤‡∏´‡πå'].fillna(mean_hour_per_week).round(1)
```

```python
df_cleaned = df_cleaned[(df_cleaned['‡∏≠‡∏≤‡∏¢‡∏∏'] >= lower_bound_age) & (df_cleaned['‡∏≠‡∏≤‡∏¢‡∏∏'] <= upper_bound_age)]
df_cleaned = df_cleaned[(df_cleaned['GPA'] >= lower_bound_gpa) & (df_cleaned['GPA'] <= upper_bound_gpa)]
```

#‡∏ó‡∏≥‡πÉ‡∏´‡πâ‡∏ä‡πà‡∏≠‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏• ‡πÄ‡∏û‡∏®, ‡∏Ñ‡∏ì‡∏∞ ‡πÉ‡∏´‡πâ‡πÄ‡∏õ‡πá‡∏ô‡∏°‡∏≤‡∏ï‡∏£‡∏ê‡∏≤‡∏ô

```python
def standardize_faculty(name):
    if name in ['‡∏ß‡∏¥‡∏ó‡∏¢‡∏≤‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå', '‡∏ß‡∏¥‡∏ó‡∏¢‡πå', 'Science', '‡∏Ñ‡∏ì‡∏∞‡∏ß‡∏¥‡∏ó‡∏¢‡∏≤‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå','Sci']:
        return '‡∏ß‡∏¥‡∏ó‡∏¢‡∏≤‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå'
    elif name in ['‡∏®‡∏¥‡∏•‡∏õ‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå', '‡∏®‡∏¥‡∏•‡∏õ‡πå', 'Arts','Liberal Arts']:
        return '‡∏®‡∏¥‡∏•‡∏õ‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå'
    elif name in ['‡∏ß‡∏¥‡∏®‡∏ß‡∏Å‡∏£‡∏£‡∏°‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå','‡∏ß‡∏¥‡∏®‡∏ß‡∏∞','‡∏ß‡∏¥‡∏®‡∏ß‡∏Ø','Engineering','‡∏ß‡∏¥‡∏®‡∏ß‡∏Å‡∏£‡∏£‡∏°']:
        return '‡∏ß‡∏¥‡∏®‡∏ß‡∏Å‡∏£‡∏£‡∏°‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå'
    else:
        return name
```

```python
def standardize_sex(sex):
    if sex in ['‡∏ä‡∏≤‡∏¢', '‡∏ä', 'M']:
        return '‡∏ä‡∏≤‡∏¢'
    elif sex in ['‡∏´‡∏ç‡∏¥‡∏á', '‡∏ç', 'Female','F']:
        return '‡∏´‡∏ç‡∏¥‡∏á'
    else:
        return sex
```

```python
df_cleaned['‡∏Ñ‡∏ì‡∏∞'] = df_cleaned['‡∏Ñ‡∏ì‡∏∞'].apply(standardize_faculty)
df_cleaned['‡πÄ‡∏û‡∏®'] = df_cleaned['‡πÄ‡∏û‡∏®'].apply(standardize_sex)
```
